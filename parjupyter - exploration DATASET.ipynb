```python
import pandas as pd
import json  

cheminjson = 'DATASET - COMMUNES (v.elecPW2024-2019).json'

with open(cheminjson, 'r', encoding='utf-8') as f:
    data = json.load(f)

rows = []
for commune, elections in data.items():
    for election, partis in elections.items():
        for parti, result in partis.items():
            rows.append([commune, election, parti, result])


df = pd.DataFrame(rows, columns=["Commune", "Election", "Parti", "Résultat"])


df_pivot = df.pivot_table(
    index="Commune",
    columns=["Election", "Parti"],
    values="Résultat",
    aggfunc=lambda x: x
)


display(df_pivot)
```
